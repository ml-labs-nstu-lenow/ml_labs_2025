## Обучение модели

Разделим данные на две выборки

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
    df.drop('class_type', axis=1),
    df['class_type'],
    test_size=0.3,
    random_state=42,
    stratify=df['class_type'],
)
```

Определим параметры для GridSearchCV и модели для классификации.

```python
param_grid_lr = {
    'estimator__penalty': ['l2'],
    'estimator__C': [0.1, 1, 10],
    'estimator__solver': ['lbfgs', 'sag', 'saga'],
    'estimator__max_iter': [1000],
}
```

```python
line_model = sklearn.linear_model.LogisticRegression()
```

```python
param_grid_lasso = {
    'estimator__penalty': ['l1'],
    'estimator__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'estimator__solver': ['saga'],
    'estimator__max_iter': [5000, 10000],
    'estimator__tol': [1e-3, 1e-4],
    'estimator__class_weight': [None, 'balanced'],
    'estimator__warm_start': [True],
}
```

```python
lasso_model =  sklearn.linear_model.LogisticRegression(penalty='l1')
```

```python
param_grid_ridge = {
    'estimator__penalty': ['l2'],
    'estimator__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'estimator__solver': ['newton-cg', 'lbfgs', 'sag', 'saga'],
    'estimator__max_iter': [100, 500, 1000],
    'estimator__class_weight': [None, 'balanced'],
}
```

```python
ridge_model = sklearn.linear_model.LogisticRegression(penalty='l2')
```

```python
param_grid_tree = {
    'estimator__criterion': ['gini', 'entropy'],
    'estimator__max_depth': [3, 5, 7, 10, None],
    'estimator__min_samples_split': [2, 5, 10],
    'estimator__min_samples_leaf': [1, 2, 4],
    'estimator__max_features': ['sqrt', 'log2', None],
    'estimator__class_weight': [None, 'balanced'],
}
```

```python
tree_model = sklearn.tree.DecisionTreeClassifier()
```

Определим лучшие парамтеры для каждой модели.

```python
scoring_metric = 'f1_macro'
```

```python
line_gs = sklearn.model_selection.GridSearchCV(
    sklearn.multiclass.OneVsRestClassifier(line_model),
    param_grid_lr,
    scoring=scoring_metric,
    verbose=0,
)
line_gs.fit(x_train, y_train)
```

```python
lasso_gs = sklearn.model_selection.GridSearchCV(
    sklearn.multiclass.OneVsRestClassifier(lasso_model),
    param_grid_lasso,
    scoring=scoring_metric,
    cv=5,
    verbose=2,
    n_jobs=-1
)
lasso_gs.fit(x_train, y_train)
```

```python
ridge_gs = sklearn.model_selection.GridSearchCV(
    sklearn.multiclass.OneVsRestClassifier(ridge_model),
    param_grid_ridge,
    scoring=scoring_metric,
    cv=5,
    verbose=0,
    n_jobs=-1,
)
ridge_gs.fit(x_train, y_train)
```

```python
tree_gs = sklearn.model_selection.GridSearchCV(
    sklearn.multiclass.OneVsRestClassifier(tree_model),
    param_grid_tree,
    scoring=scoring_metric,
    verbose=0,
)
tree_gs.fit(x_train, y_train)
```

```python
y_test_bin = sklearn.preprocessing.label_binarize(y_test, classes=np.unique(y_test))
n_classes = y_test_bin.shape[1]

fig, axes = plt.subplots(2, 2, figsize=(18, 14))
models = [('Linear', line_gs), ('Lasso', lasso_gs),
          ('Ridge', ridge_gs), ('Tree', tree_gs)]

for ax, (name, model) in zip(axes.ravel(), models):
      y_score = model.best_estimator_.predict_proba(x_test)
      y_score = np.nan_to_num(y_score, nan=0.0, posinf=1.0, neginf=0.0)

      ap_scores = []
      for i in range(n_classes):
          precision, recall, _ = precision_recall_curve(y_test_bin[:, i], y_score[:, i])
          ap = average_precision_score(y_test_bin[:, i], y_score[:, i])
          ap_scores.append(ap)
          ax.plot(recall, precision, lw=2, label=f'Class {i+1} (AP={ap:.2f})')

      precision_micro, recall_micro, _ = precision_recall_curve(
          y_test_bin.ravel(), y_score.ravel())
      ap_micro = average_precision_score(y_test_bin, y_score, average='micro')
      ax.plot(recall_micro, precision_micro, 'k--',
              label=f'Micro-avg (AP={ap_micro:.2f})')

      ax.set_title(f'{name} Model', fontsize=12)
      ax.set_xlabel('Recall')
      ax.set_ylabel('Precision')
      ax.legend(loc="lower left", fontsize=8)
      ax.grid(True)

plt.tight_layout()
plt.show()
```

Как и ожидалось модели с и без L1, L2 почти не имеют разлиций, так как в данные не осталось высоко скореллированных значений.
