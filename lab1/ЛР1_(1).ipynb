```python
top = df1[:10].index.tolist()
df_top = df[df['SUBDIVISION'].isin(top)]

months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
df_top = df_top.drop(['Jan-Feb','Mar-May','Jun-Sep','Oct-Dec','ANNUAL'],axis=1)
df_top = pd.melt(df_top, id_vars=['SUBDIVISION', 'YEAR'],
                           value_vars=months,
                           var_name='Month',
                           value_name='Rainfall')

plts = sns.catplot(
    data=df_top,
    x='Month',
    y='Rainfall',
    col='SUBDIVISION',
    kind='box',
    col_wrap=5,
    height=5,
    aspect=1.2,
    sharey=False,
)
plts.set_ylabels(' ()')
plts.tick_params(axis='x', labelbottom=True)
plt.tight_layout()
plt.show()
```
