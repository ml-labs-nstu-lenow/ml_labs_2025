4. Подсчитайте для трёх выбранных авиакомпаний: количество рейсов, количество отменённых рейсов, количество перенаправленных рейсов. По каждому типу постройте histplot.

```python
import matplotlib.pyplot as plt


target_airlines = ['UA', 'AA', 'US']
filtered_df = df[df['AIRLINE'].isin(target_airlines)].copy()


for airline in target_airlines:
    airline_data = filtered_df[filtered_df['AIRLINE'] == airline]


    total = len(airline_data)
    cancelled = airline_data['CANCELLED'].sum()
    diverted = airline_data['DIVERTED'].sum()


    plt.figure(figsize=(7, 4), facecolor='white')
    bars = plt.bar(['Всего рейсов', 'Отмененные', 'Перенапрвленные'],
                  [total, cancelled, diverted],
                  color=['#1f77b4', '#ff7f0e', '#2ca02c'])


    plt.title(f'Рейсы {airline}')
    plt.gca().set_facecolor('white')


    for bar in bars:
        h = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, h, f'{int(h)}',
                ha='center', va='bottom', fontsize=10)

    plt.tight_layout()
    plt.show()
```

5. Сформируйте новый признак скорости полёта для каждого рейса. Рассчитайте среднюю скорость полёта по всем рейсам для трёх выбранных авиакомпаний и выведите её на histplot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['SPEED'] = df['DISTANCE'] / (df['AIR_TIME'] / 60)

selected_airlines = ['UA', 'AA', 'US']

for airline in selected_airlines:
    airline_data = df[df['AIRLINE'] == airline]

    avg_speed = airline_data['SPEED'].mean()

    plt.figure(figsize=(10, 6))
    sns.histplot(airline_data['SPEED'], bins=20, kde=True)

    plt.title(f"Распределение скорости полёта для авиакомпании {airline}")
    plt.xlabel("Скорость (миль/ч)")
    plt.ylabel("Частота")
    plt.show()

    print(f"Средняя скорость полёта для авиакомпании {airline}: {avg_speed:.2f} миль/ч")
```

6. Возьмите 10 дней из общей выборки. Визуализируйте тепловую карту (heatplot) частоты отмены рейсов. По одной оси – дни, по другой оси – рейс (для двух аэропортов).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df['DATE'] = pd.to_datetime(df[['YEAR', 'MONTH', 'DAY']])


days = df['DATE'].drop_duplicates().sample(10, random_state=1)
airports = df['ORIGIN_AIRPORT'].value_counts().head(2).index


small_df = df[
    df['DATE'].isin(days) &
    df['ORIGIN_AIRPORT'].isin(airports)
]


heat_data = small_df.groupby(['DATE', 'ORIGIN_AIRPORT'])['CANCELLED'].mean().unstack()


plt.figure(figsize=(12, 6))
sns.heatmap(heat_data.T,
           cmap='YlOrRd',
           annot=True,
           fmt='.1%',
           linewidths=0.5,
           cbar_kws={'label': 'Доля отмен'})

plt.title('Тепловая карта отмен рейсов по дням и аэропортам', pad=20)
plt.xlabel('Дата')
plt.ylabel('Аэропорт отправления')
plt.xticks(rotation=45)

date_labels = heat_data.index.strftime('%Y-%m-%d')
plt.xticks(ticks=range(len(date_labels)), labels=date_labels)

plt.tight_layout()
plt.show()
```

7. Сформируйте новый датасет по авиакомпаниям с признаками: среднее время задержки отправки и среднее время прибытия.

```python
delays = df.groupby('AIRLINE').agg({
    'DEPARTURE_DELAY': 'mean',
    'ARRIVAL_DELAY': 'mean'
}).reset_index()

print(delays.sort_values('DEPARTURE_DELAY', ascending=False))
```

8. Постройте два histplot по задержкам отправления и прибытия для топа 5 компаний.

```python
import seaborn as sns
import matplotlib.pyplot as plt

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8']

top5_arrival = delays.sort_values('ARRIVAL_DELAY', ascending=False).head(5)

plt.figure(figsize=(12, 6))
ax = sns.barplot(data=top5_arrival, x='AIRLINE', y='ARRIVAL_DELAY', hue='AIRLINE',
                 order=top5_arrival['AIRLINE'], palette=colors, legend=False)

for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f} мин",
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10,
                xytext=(0, 7),
                textcoords='offset points')

plt.title("Топ-5 авиакомпаний с наибольшими задержками прибытия", pad=20, fontsize=14)
plt.xlabel('Авиакомпания', fontsize=12)
plt.ylabel('Средняя задержка (минуты)', fontsize=12)
sns.despine(left=True)
plt.show()


top5_departure = delays.sort_values('DEPARTURE_DELAY', ascending=False).head(5)

plt.figure(figsize=(12, 6))
ax = sns.barplot(data=top5_departure, x='AIRLINE', y='DEPARTURE_DELAY', hue='AIRLINE',
                 order=top5_departure['AIRLINE'], palette=colors[::-1], legend=False)


for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f} мин",
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10,
                xytext=(0, 7),
                textcoords='offset points')

plt.title("Топ-5 авиакомпаний с наибольшими задержками отправления", pad=20, fontsize=14)
plt.xlabel('Авиакомпания', fontsize=12)
plt.ylabel('Средняя задержка (минуты)', fontsize=12)
sns.despine(left=True)
plt.show()
```

9. Определите авиакомпанию с максимальными задержками рейсов по отправке и прибытию (самый ярый опаздун).

```python
delays['TOTAL_DELAY'] = delays['DEPARTURE_DELAY'] + delays['ARRIVAL_DELAY']

max_delay = delays.loc[delays['TOTAL_DELAY'].idxmax()]

print(f"Авиакомпания по самым частым задержкам: {max_delay['AIRLINE']}")
print(f"- Средняя задержка отправления: {max_delay['DEPARTURE_DELAY']:.2f} мин")
print(f"- Средняя задержка прибытия: {max_delay['ARRIVAL_DELAY']:.2f} мин")
```
