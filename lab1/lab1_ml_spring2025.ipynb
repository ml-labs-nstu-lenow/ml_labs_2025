Заполняем пропуски в скоростях через knn imputer с таргетом на рад. энергию

```python
from sklearn.impute import KNNImputer
X = df[['Velocity (km/s)', 'vx', 'vy', 'vz', 'Total Radiated Energy (J)']]
imputer = KNNImputer(n_neighbors=2)
X_imputed = imputer.fit_transform(X)

df_imputed = pd.DataFrame(X_imputed, columns=X.columns)
df_final = df.fillna(df_imputed)
df_final.isnull().sum().sort_values(ascending=False)
```

место максимальной концентрации метеоритов

```python
def parse_coord(coord_str):
    if pd.isna(coord_str) or not isinstance(coord_str, str):
        return None
    value = float(coord_str[:-1])
    direction = coord_str[-1].upper()

    if direction in ['N', 'E']:
        return value
    elif direction in ['S', 'W']:
          return -value
    else:
        return None
    return None


df_final['Latitude_dec'] = df_final['Latitude (deg.)'].apply(parse_coord)
df_final['Longitude_dec'] = df_final['Longitude (deg.)'].apply(parse_coord)


# Группируем по округленным координатам (шаг 10°)
df_final['Lat_rounded'] = (df_final['Latitude_dec'] // 10) * 10
df_final['Lon_rounded'] = (df_final['Longitude_dec'] // 10) * 10
concentration = df_final.groupby(['Lat_rounded', 'Lon_rounded']).size().reset_index(name='Count')

# Находим регион с максимальной концентрацией
max_concentration = concentration.loc[concentration['Count'].idxmax()]
max_concentration
```

Вычисляем корреляцию Пирсона и Спирмена между признаками Altitude и Total Radiated Energy

```python
from scipy import stats
import seaborn as sns
import numpy as np

df_cleaned = df[['Altitude (km)', 'Total Radiated Energy (J)']].dropna()

pearson_corr, pearson_p = stats.pearsonr(df_cleaned['Altitude (km)'],
                                        df_cleaned['Total Radiated Energy (J)'])
spearman_corr, spearman_p = stats.spearmanr(df_cleaned['Altitude (km)'],
                                           df_cleaned['Total Radiated Energy (J)'])

print(f"Корреляция Пирсона: {pearson_corr:.3f} (p-value = {pearson_p:.3e})")
print(f"Корреляция Спирмена: {spearman_corr:.3f} (p-value = {spearman_p:.3e})")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_cleaned['Altitude (km)'], bins=30, kde=True, color='skyblue')
plt.title('Распределение высот (Altitude)')
plt.xlabel('Высота (км)')

plt.subplot(1, 2, 2)
sns.histplot(np.log10(df_cleaned['Total Radiated Energy (J)']), bins=30, kde=True, color='salmon')
plt.title('Распределение энергии (log10)')
plt.xlabel('log10(Энергия, Дж)')

plt.tight_layout()
plt.show()
```

Вывод:
Корреляция Пирсона: нет статистически значимых доказательств линейной связи между переменными.
Корреляция Спирмена: Есть статистически значимая (но слабая) отрицательная монотонная зависимость.
