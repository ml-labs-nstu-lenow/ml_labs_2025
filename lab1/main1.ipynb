```python
import matplotlib.pyplot as plt
```

```python
unique_states = df_cleaned['SUBDIVISION'].unique()
num_states = len(unique_states)

# Настройки для подграфиков
cols = 3  # количество графиков в строке
rows = (num_states + cols - 1) // cols

# Создаем сетку
plt.figure(figsize=(15, rows*3))

for i, state in enumerate(unique_states):
    plt.subplot(rows, cols, i+1)
    state_data = df_cleaned[df_cleaned['SUBDIVISION'] == state]
    plt.plot(state_data['YEAR'], state_data['ANNUAL'])
    plt.title(state, fontsize=13)
    plt.xlabel('Год')
    plt.ylabel('Осадки (мм)')

plt.tight_layout(rect=[0, 0, 1, 1])

plt.show()
```

```python
months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']


plt.figure(figsize=(8, 5))

for month in months:
    (df_cleaned.groupby('YEAR')[month].sum()
              .cumsum()
              .plot(label=month))

plt.xlabel('Год')
plt.ylabel('Накопленные осадки (мм)')
plt.title('Накопленные осадки по месяцам за годы')
plt.show()
```

```python
unique_states = df_cleaned['SUBDIVISION'].unique()
num_states = len(unique_states)

cols = 2
rows = (num_states + cols - 1) // cols

plt.figure(figsize=(15, rows*8))

for i, state in enumerate(unique_states):
    plt.subplot(rows, cols, i+1)
    state_data = df_cleaned[df_cleaned['SUBDIVISION'] == state]
    
    # Строим boxplot для осадков по месяцам
    sns.boxplot(data=state_data[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']])
    
    plt.title(f'Осадки в {state}', fontsize=13)
    plt.xlabel('Месяц')
    plt.ylabel('Осадки (мм)')

plt.tight_layout(rect=[0, 0, 1, 1])

plt.show()
```

```python
mean_by_state = df_cleaned.groupby('SUBDIVISION')['ANNUAL'].sum()
mean_by_state = mean_by_state.sort_values()
print(mean_by_state)

max_mean = mean_by_state.max()
min_mean = mean_by_state.min()


state_max = mean_by_state[mean_by_state == max_mean].index[0]
state_min = mean_by_state[mean_by_state == min_mean].index[0]
print(state_max, max_mean)
print(state_min, min_mean)
```

Постройте гистограммы для указанных признаков. Оцените значение p-value.

```python
!uv pip install scipy
```

```python
from scipy.stats import pearsonr, spearmanr
```

```python
pearson_corr, pearson_p_value = pearsonr(df_cleaned['NOV'], df_cleaned['MAR'])

spearman_corr, spearman_p_value = spearmanr(df_cleaned['NOV'], df_cleaned['MAR'])

print(pearson_corr, spearman_corr)
print(pearson_p_value, spearman_p_value)
```
